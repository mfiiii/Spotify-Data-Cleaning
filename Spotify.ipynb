```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#xóa các cột Unnamed: 0
df = df.drop("Unnamed: 0", axis = 1)
```

```python
a = df.groupby("artist", as_index=False)["song_title"].count().sort_values(by="song_title", ascending = False).head()
a
#Các hàng a sẽ được sắp xếp theo thứ tự giảm dần của số lượng bài hát, cho bạn biết top 5 nghệ sĩ có nhiều bài hát nhất trong dữ liệu.
```

```python
sns.barplot(x="song_title", y="artist", hue="artist", data=a, palette="ocean", legend=False)
```

```python
b = df[["loudness", "song_title"]].sort_values(by="loudness", ascending =True).head()
b
#xếp các bài hát theo âm thanh thấp đến cao
```

```python
sns.barplot(x="loudness", y="song_title", hue="song_title", data=b, palette="gist_heat", legend=False)
```

```python
c = df[["danceability", "song_title", "artist"]].sort_values(by = "danceability", ascending = False).head()
c
```

```python
sns.barplot(y="danceability", x="artist", hue="artist", data=c, palette="flare", legend=True)
```

```python
d = df[["instrumentalness", "song_title", "artist"]].sort_values(by = "instrumentalness", ascending = False).head()
d
```

```python
sns.barplot(y="instrumentalness", x="artist", hue="artist", data=d, palette="dark", legend=True)
#các nghệ sĩ có nhiều bài hát sử dụng nhiều nhạc cụ hơn 
```

```python
plt.figure(figsize=(10,5))#10inch rộng và 5 inch cao 
plt.pie(x="instrumentalness", data = d, autopct = "%1.3f%%", labels = d.song_title)
plt.title("Top 5 instrumental tracks")
plt.show()
```
